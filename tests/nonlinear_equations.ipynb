# Решение нелинейных уравнений

```python
import sys
import os

import numpy as np
from scipy.integrate import quad
from matplotlib import pyplot as plt

# Добавляем путь к корневой директории проекта
sys.path.append(os.path.abspath(os.path.join(os.getcwd(), '..')))

from mathmod import nonlinear_equations
```

```python
eps = 10e-6 # устанавливаем погрешность
```

```python
def f(x): # тестовая функция
    return x ** 4 - 7 * x ** 2 + 10

def df(x):
    return 4 * x ** 3 - 7 * 2 * x
```

```python
x = np.linspace(-3, 1, 100)
plt.figure(figsize=(10, 8))
plt.title("График функции f(x)")
plt.plot(x, f(x))
plt.grid()
plt.show()
```

```python
x_minus_1 = -3
x_1 = -2

x_1_secant, n_secant = nonlinear_equations.secant(f, x_minus_1, x_1, eps)
print(f"1-ы корень, вычисленный методом секущих: {x_1_secant}, количество итераций: {n_secant}")
```

```python
a = -3
b = -2
x_1_bisection, n_bisection = nonlinear_equations.bisection(f, a, b, eps)
print(f"1-ы корень, вычисленный методом бисекций: {x_1_bisection}, количество итераций: {n_bisection}")
```

```python
a = -3
b = -2
x_1_false_position, n_false_position = nonlinear_equations.false_position(f, a, b, eps)
print(f"1-ы корень, вычисленный методом бисекций: {x_1_false_position}, количество итераций: {n_false_position}")
```

```python
x_0_newton = -3
x_1_newton, n_newton = nonlinear_equations.newton(f, df, x_0_newton, eps)
print(f"1-ы корень, вычисленный методом Ньютона: {x_1_newton}, количество итераций: {n_newton}")
```

```python
x_0_simplifed_newton = -3
x_1_simplifed_newton, n_simplifed_newton = nonlinear_equations.simplified_newton(f, df, x_0_simplifed_newton, eps)
print(f"1-ы корень, вычисленный упрощенным методом Ньютона: {x_1_simplifed_newton}, количество итераций: {n_simplifed_newton}")
```
